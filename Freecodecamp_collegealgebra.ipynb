```python
#Solve and graph. Factor to change from "0 =" to "y =" preparing to graph
from sympy import *
from sympy.plotting import plot
from sympy import sqrt

var('x y')

# First equation set equal to zero, ready to solve
first = -x**2 - y + 10
#first = sqrt(x) - y

#Second equation set equal to zero, ready to solve
second = 2*x**2 - 2*y - 4
#second = -x + 5 - y

# Solve - can be linear or nonlinear equations
solution = nonlinsolve([first, second], (x, y))
for a in range(len(solution.args)):
    x_solution = solution.args[a][0]
    y_solution = solution.args[a][1]
    print("Solution = (", x_solution, ",", y_solution, ")")

# Sympy syntax for equation equal to zero, ready to factor
y_first = Eq(first,0)

# Sympy solve for y
y1 = solve(y_first,y)

# Same two steps for second equation
y_second = Eq(second,0)
y2 = solve(y_second,y)

# Show factored results
print("y = ", y1[0])
print("y = ", y2[0])

# Plot solution
x = symbols('x')
xmin = -10
xmax = 10
plot(y1[0], y2[0], (x,xmin,xmax))

```

```python

%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
import math

# All graphing happens in this function
def f(a,b,c):
    xmin = -10
    xmax = 10
    ymin = -10
    ymax = 10
    points = 10*(xmax-xmin)
    x = np.linspace(xmin,xmax,points)

    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'b') # x axis
    plt.plot([0,0],[ymin,ymax], 'b') # y axis

    # Parabola
    y1 = a*x**2 + b*x + c
    plt.plot(x, y1)

    # Vertex
    vx = -b/(2*a)
    vy = a*(vx**2) + b*vx + c
    plt.plot([vx],[vy], 'ro')

    # Roots
    d = b**2 - 4*a*c
    if d>=0:
        root_1 = (-b + math.sqrt(d))/(2*a)
        root_2 = (-b - math.sqrt(d))/(2*a)
        plt.plot([root_1, root_2],[0,0], 'go')

    # Set the equation as the title
    sa = str(a)
    sb = str(b)
    sc = str(c)
    h1 = "y = ", sa, "x**2 + ", sb, "x + ", sc
    h2 = ""
    for w in h1:
        h2 = h2 + w

    plt.title(h2)
    plt.show()

# Interactive creates sliders for a, b, and c values
interactive_plot = interactive(f, a=(1, 9), b=(-9,9), c=(-9,9))
interactive_plot

```

```python

import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve, N
from sympy.solvers import solve

# Define the function to graph one or more functions
def graph_functions(functions, x_range):
    x = np.linspace(x_range[0], x_range[1], 400)
    for func in functions:
        y = eval(func)
        plt.plot(x, y, label=f'y={func}')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()

# Define the function to create a table of (x, y) values
def create_table(function, x_values):
    table = []
    for x in x_values:
        y = eval(function)
        table.append((x, y))
    return table

# Define the function to shade above or below the line
def shade_area(function, x_range, shade_above=True):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = eval(function)
    plt.plot(x, y, label=f'y={function}')
    if shade_above:
        plt.fill_between(x, y, y2=max(y), alpha=0.3)
    else:
        plt.fill_between(x, y, y2=min(y), alpha=0.3)
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()

# Define the function to solve and graph a system of equations
def solve_and_graph_system(eq1, eq2, x_range):
    x, y = symbols('x y')
    eq1_sympy = Eq(eval(eq1.replace('=', '-(') + ')'), 0)
    eq2_sympy = Eq(eval(eq2.replace('=', '-(') + ')'), 0)
    solution = solve((eq1_sympy, eq2_sympy), (x, y))
    x_vals = np.linspace(x_range[0], x_range[1], 400)
    y1_vals = [N(eq1_sympy.rhs.subs(x, val)) for val in x_vals]
    y2_vals = [N(eq2_sympy.rhs.subs(x, val)) for val in x_vals]
    plt.plot(x_vals, y1_vals, label=f'{eq1}')
    plt.plot(x_vals, y2_vals, label=f'{eq2}')
    if solution:
        plt.plot(solution[x], solution[y], 'ro')
        plt.annotate(f'Solution: ({solution[x]}, {solution[y]})', (solution[x], solution[y]))
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()

# Define the function to zoom in or out on a graph
def zoom_graph(function, x_range, zoom_factor):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = eval(function)
    plt.plot(x, y, label=f'y={function}')
    plt.xlim([x_range[0] / zoom_factor, x_range[1] / zoom_factor])
    plt.ylim([min(y) / zoom_factor, max(y) / zoom_factor])
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()

# Define the function to solve quadratic equations
def solve_quadratic(a, b, c):
    x = symbols('x')
    equation = Eq(a * x**2 + b * x + c, 0)
    solutions = solve(equation, x)
    return solutions

# Example usage
if __name__ == "__main__":
    # Graphing functions example
    functions = ["np.sin(x)", "np.cos(x)", "x**2"]
    x_range = [-10, 10]
    graph_functions(functions, x_range)

    # Creating a table of values example
    function = "np.sin(x)"
    x_values = np.linspace(-10, 10, 10)
    table = create_table(function, x_values)
    print("Table of (x, y) values:")
    for row in table:
        print(row)

    # Shading above a line example
    shade_area("np.sin(x)", [-10, 10], shade_above=True)

    # Solving and graphing a system of equations example
    eq1 = "2*x + 3*y - 6"
    eq2 = "4*x - 2*y - 4"
    solve_and_graph_system(eq1, eq2, [-10, 10])

    # Zooming in on a graph example
    zoom_graph("np.sin(x)", [-10, 10], zoom_factor=2)

    # Solving a quadratic equation example
    a, b, c = 1, -3, 2  # x^2 - 3x + 2 = 0
    solutions = solve_quadratic(a, b, c)
    print(f"Solutions to the quadratic equation: {solutions}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from io import BytesIO
from urllib.request import urlopen
from IPython.display import display
import ipywidgets as widgets
from google.colab import files
from ipywidgets import interact

# Function to load CSV from local computer
def upload_csv():
    uploaded = files.upload()
    for fn in uploaded.keys():
        return pd.read_csv(BytesIO(uploaded[fn]))

# Function to load CSV from a URL
def load_csv_from_url(url):
    return pd.read_csv(url)

# Function to print headings and the first two rows
def print_headings_and_first_two_rows(df):
    print("Headings:")
    print(df.columns)
    print("\nFirst two rows:")
    print(df.head(2))

# Function to store column names as a list
def get_column_names(df):
    return df.columns.tolist()

# Function to plot data
def plot_data(df, column1, column2, plot_type='scatter'):
    x = df[column1].to_numpy()
    y = df[column2].to_numpy()

    plt.figure(figsize=(10, 6))
    if plot_type == 'scatter':
        plt.scatter(x, y)
    elif plot_type == 'line':
        plt.plot(x, y)
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.title(f'{plot_type.capitalize()} plot of {column1} vs {column2}')
    plt.grid(True)
    plt.show()

# Function to interactively choose columns and plot type
def interactive_plot(df):
    column_names = get_column_names(df)

    column1_widget = widgets.Dropdown(
        options=column_names,
        description='Column 1:',
        disabled=False,
    )

    column2_widget = widgets.Dropdown(
        options=column_names,
        description='Column 2:',
        disabled=False,
    )

    plot_type_widget = widgets.Dropdown(
        options=['scatter', 'line'],
        description='Plot Type:',
        disabled=False,
    )

    def update_plot(column1, column2, plot_type):
        plot_data(df, column1, column2, plot_type)

    interact(update_plot, column1=column1_widget, column2=column2_widget, plot_type=plot_type_widget)

# Main function
def main():
    # Load CSV from the provided URL
    url = 'https://raw.githubusercontent.com/AlexTheAnalyst/PandasYouTubeSeries/main/Flavors.csv'
    df = load_csv_from_url(url)

    # Print headings and first two rows
    print_headings_and_first_two_rows(df)

    # Interactive plot
    interactive_plot(df)

# Running the main function
if __name__ == "__main__":
    main()
```

Scatter Plot Game

```python
import matplotlib.pyplot as plt
import numpy as np

def scatter_plot_game():
    # Generate random points
    num_points = 5
    x_values = np.random.randint(-10, 10, num_points)
    y_values = np.random.randint(-10, 10, num_points)

    # Plot the points
    plt.scatter(x_values, y_values)
    plt.xlim(-15, 15)
    plt.ylim(-15, 15)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Scatter Plot Game')
    plt.grid(True)
    plt.show()

    # Ask the user to guess the points
    correct_guesses = 0
    for i in range(num_points):
        guess_x = int(input(f"Enter x coordinate for point {i + 1}: "))
        guess_y = int(input(f"Enter y coordinate for point {i + 1}: "))
        if guess_x == x_values[i] and guess_y == y_values[i]:
            correct_guesses += 1

    print(f"You guessed {correct_guesses} out of {num_points} points correctly!")

scatter_plot_game()
```

Algebra Practice Game

```python
import random

def algebra_practice_game():
    num_problems = 5
    correct_answers = 0

    for _ in range(num_problems):
        problem_type = random.choice(['one-step', 'two-step'])
        if problem_type == 'one-step':
            a = random.randint(-10, 10)
            b = random.randint(-10, 10)
            print(f"Solve for x: {a}x = {b}")
            answer = b / a
        else:
            a = random.randint(-10, 10)
            b = random.randint(-10, 10)
            c = random.randint(-10, 10)
            print(f"Solve for x: {a}x + {b} = {c}")
            answer = (c - b) / a

        user_answer = float(input("Your answer: "))
        if user_answer == answer:
            print("Correct!")
            correct_answers += 1
        else:
            print(f"Incorrect. The correct answer is {answer}")

    print(f"You answered {correct_answers} out of {num_problems} problems correctly!")

algebra_practice_game()
```

Projectile Game

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from IPython.display import display

def projectile_path(v0, angle, h, d):
    g = 9.8
    angle = np.radians(angle)
    t_flight = (2 * v0 * np.sin(angle)) / g
    t = np.linspace(0, t_flight, num=500)
    x = v0 * t * np.cos(angle)
    y = v0 * t * np.sin(angle) - 0.5 * g * t**2

    plt.plot(x, y)
    plt.axvline(x=d, color='r', linestyle='--', label='Wall')
    plt.axhline(y=h, color='g', linestyle='--', label='Height')
    plt.xlim(0, 2*d)
    plt.ylim(0, max(y)*1.1)
    plt.xlabel('Distance (m)')
    plt.ylabel('Height (m)')
    plt.title('Projectile Motion')
    plt.legend()
    plt.grid(True)
    plt.show()

def interactive_projectile_game():
    v0_slider = (1, 100, 1)
    angle_slider = (1, 89, 1)
    h_slider = (1, 50, 1)
    d_slider = (1, 100, 1)
    interact(projectile_path, v0=v0_slider, angle=angle_slider, h=h_slider, d=d_slider)

interactive_projectile_game()
```

Financial Calculator Implementation

```python
import math

# 1. Calculate annuity with monthly or continuous growth
def calculate_annuity(PMT, r, n, growth_type='monthly'):
    if growth_type == 'monthly':
        future_value = PMT * (((1 + r/n)**(n*n) - 1) / (r/n))
    elif growth_type == 'continuous':
        future_value = PMT * (math.e**(r*n) - 1) / r
    else:
        raise ValueError("Invalid growth type. Use 'monthly' or 'continuous'.")
    return future_value

# 2. Calculate monthly mortgage payment
def calculate_mortgage_payment(P, r, n):
    monthly_rate = r / 12
    num_payments = n * 12
    monthly_payment = (P * monthly_rate * (1 + monthly_rate)**num_payments) / ((1 + monthly_rate)**num_payments - 1)
    return monthly_payment

# 3. Estimate retirement investment balance
def estimate_retirement_balance(PV, PMT, r, n):
    future_value = PV * (1 + r)**n + PMT * (((1 + r)**n - 1) / r)
    return future_value

# 4. Determine how long until an amount doubles, given the rate
def time_to_double(rate):
    doubling_time = math.log(2) / math.log(1 + rate)
    return doubling_time

# 5. Solve logarithmic equations
def solve_log_equation(a, b, c):
    # Solves equations of the form a * log(x, base) = b + c
    x = 10 ** ((b + c) / a)
    return x

# 6. Convert to (and from) scientific notation
def to_scientific_notation(number):
    return f"{number:.2e}"

def from_scientific_notation(sci_notation):
    return float(sci_notation)

# Example usage of the functions
if __name__ == "__main__":
    # Annuity Calculation
    PMT = 1000  # Monthly payment
    r = 0.05  # Annual interest rate
    n = 10  # Number of years
    annuity_value_monthly = calculate_annuity(PMT, r, n, 'monthly')
    annuity_value_continuous = calculate_annuity(PMT, r, n, 'continuous')
    print(f"Annuity Value (Monthly Growth): {annuity_value_monthly}")
    print(f"Annuity Value (Continuous Growth): {annuity_value_continuous}")

    # Mortgage Payment Calculation
    P = 300000  # Principal loan amount
    r = 0.04  # Annual interest rate
    n = 30  # Number of years
    mortgage_payment = calculate_mortgage_payment(P, r, n)
    print(f"Monthly Mortgage Payment: {mortgage_payment}")

    # Retirement Investment Balance Estimation
    PV = 50000  # Present value of the investment
    PMT = 1000  # Monthly contribution
    r = 0.06  # Annual interest rate
    n = 20  # Number of years
    retirement_balance = estimate_retirement_balance(PV, PMT, r, n)
    print(f"Retirement Investment Balance: {retirement_balance}")

    # Time to Double Calculation
    rate = 0.07  # Annual growth rate
    doubling_time = time_to_double(rate)
    print(f"Time to Double: {doubling_time} years")

    # Solve Logarithmic Equation
    a = 2
    b = 4
    c = 3
    solution = solve_log_equation(a, b, c)
    print(f"Solution to Logarithmic Equation: {solution}")

    # Scientific Notation Conversion
    number = 123456789
    sci_notation = to_scientific_notation(number)
    print(f"Scientific Notation: {sci_notation}")
    original_number = from_scientific_notation(sci_notation)
    print(f"Original Number: {original_number}")
```

Multifunction Calculator

```python
from sympy import symbols, Eq, solve, Rational
from fractions import Fraction
import math

# Function to solve proportions
def solve_proportion(a, b, c, d):
    x = symbols('x')
    proportion = Eq(a / b, c / d)
    solution = solve(proportion, x)
    return solution[0]

# Function to solve for x in equations
def solve_equation(equation):
    x = symbols('x')
    equation = Eq(equation, 0)
    solution = solve(equation, x)
    return solution

# Function to factor square roots
def factor_square_root(n):
    factors = []
    for i in range(1, int(math.sqrt(n)) + 1):
        if n % (i * i) == 0:
            factors.append(i)
    return factors

# Function to convert decimals to fractions and percents
def decimal_to_fraction_and_percent(decimal):
    fraction = Fraction(decimal).limit_denominator()
    percent = decimal * 100
    return fraction, percent

# Function to convert fractions to decimals and percents
def fraction_to_decimal_and_percent(numerator, denominator):
    fraction = Fraction(numerator, denominator)
    decimal = float(fraction)
    percent = decimal * 100
    return decimal, percent

# Function to convert percents to decimals and fractions
def percent_to_decimal_and_fraction(percent):
    decimal = percent / 100
    fraction = Fraction(decimal).limit_denominator()
    return decimal, fraction

# Example usage
if __name__ == "__main__":
    # Solve proportion example: a/b = c/d, solve for x
    a, b, c, d = 3, 4, 6, symbols('x')
    print(f"Solving proportion {a}/{b} = {c}/{d}")
    print("Solution for d:", solve_proportion(a, b, c, d))

    # Solve for x in equation example: 2x + 3 = 0
    equation = 2 * symbols('x') + 3
    print(f"Solving equation: {equation} = 0")
    print("Solution for x:", solve_equation(equation))

    # Factor square root example: 36
    n = 36
    print(f"Factors of square root of {n}")
    print("Factors:", factor_square_root(n))

    # Decimal to fraction and percent example: 0.75
    decimal = 0.75
    fraction, percent = decimal_to_fraction_and_percent(decimal)
    print(f"Decimal {decimal} as fraction and percent")
    print("Fraction:", fraction)
    print("Percent:", percent)

    # Fraction to decimal and percent example: 3/4
    numerator, denominator = 3, 4
    decimal, percent = fraction_to_decimal_and_percent(numerator, denominator)
    print(f"Fraction {numerator}/{denominator} as decimal and percent")
    print("Decimal:", decimal)
    print("Percent:", percent)

    # Percent to decimal and fraction example: 75%
    percent = 75
    decimal, fraction = percent_to_decimal_and_fraction(percent)
    print(f"Percent {percent}% as decimal and fraction")
    print("Decimal:", decimal)
    print("Fraction:", fraction)
```

Graphing Calculator


```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve, N
from sympy.solvers import solve

# Define the function to graph one or more functions
def graph_functions(functions, x_range):
    x = np.linspace(x_range[0], x_range[1], 400)
    for func in functions:
        y = eval(func)
        plt.plot(x, y, label=f'y={func}')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()

# Define the function to create a table of (x, y) values
def create_table(function, x_values):
    table = []
    for x in x_values:
        y = eval(function)
        table.append((x, y))
    return table

# Define the function to shade above or below the line
def shade_area(function, x_range, shade_above=True):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = eval(function)
    plt.plot(x, y, label=f'y={function}')
    if shade_above:
        plt.fill_between(x, y, y2=max(y), alpha=0.3)
    else:
        plt.fill_between(x, y, y2=min(y), alpha=0.3)
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()

# Define the function to solve and graph a system of equations
def solve_and_graph_system(eq1, eq2, x_range):
    x, y = symbols('x y')
    eq1_sympy = Eq(eval(eq1.replace('=', '-(') + ')'), 0)
    eq2_sympy = Eq(eval(eq2.replace('=', '-(') + ')'), 0)
    solution = solve((eq1_sympy, eq2_sympy), (x, y))
    x_vals = np.linspace(x_range[0], x_range[1], 400)
    y1_vals = [N(eq1_sympy.rhs.subs(x, val)) for val in x_vals]
    y2_vals = [N(eq2_sympy.rhs.subs(x, val)) for val in x_vals]
    plt.plot(x_vals, y1_vals, label=f'{eq1}')
    plt.plot(x_vals, y2_vals, label=f'{eq2}')
    if solution:
        plt.plot(solution[x], solution[y], 'ro')
        plt.annotate(f'Solution: ({solution[x]}, {solution[y]})', (solution[x], solution[y]))
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()

# Define the function to zoom in or out on a graph
def zoom_graph(function, x_range, zoom_factor):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = eval(function)
    plt.plot(x, y, label=f'y={function}')
    plt.xlim([x_range[0] / zoom_factor, x_range[1] / zoom_factor])
    plt.ylim([min(y) / zoom_factor, max(y) / zoom_factor])
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()

# Define the function to solve quadratic equations
def solve_quadratic(a, b, c):
    x = symbols('x')
    equation = Eq(a * x**2 + b * x + c, 0)
    solutions = solve(equation, x)
    return solutions

# Example usage
if __name__ == "__main__":
    # Graphing functions example
    functions = ["np.sin(x)", "np.cos(x)", "x**2"]
    x_range = [-10, 10]
    graph_functions(functions, x_range)

    # Creating a table of values example
    function = "np.sin(x)"
    x_values = np.linspace(-10, 10, 10)
    table = create_table(function, x_values)
    print("Table of (x, y) values:")
    for row in table:
        print(row)

    # Shading above a line example
    shade_area("np.sin(x)", [-10, 10], shade_above=True)

    # Solving and graphing a system of equations example
    eq1 = "2*x + 3*y - 6"
    eq2 = "4*x - 2*y - 4"
    solve_and_graph_system(eq1, eq2, [-10, 10])

    # Zooming in on a graph example
    zoom_graph("np.sin(x)", [-10, 10], zoom_factor=2)

    # Solving a quadratic equation example
    a, b, c = 1, -3, 2  # x^2 - 3x + 2 = 0
    solutions = solve_quadratic(a, b, c)
    print(f"Solutions to the quadratic equation: {solutions}")
```
